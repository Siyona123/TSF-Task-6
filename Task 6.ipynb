# Create the Decision Tree classifier and visualize it graphically

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
iris=datasets.load_iris()
x=iris.data
y=iris.target
```

```python
x[1]
```

```python
print(x.shape)
print(y.shape)
```

# Decision Tree

```python
from sklearn import tree
df=tree.DecisionTreeClassifier()
df.fit(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(11,11),dpi=200)
_=tree.plot_tree(df,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,fontsize=7)
```

```python
fig.savefig("Decision_Tree.png")
```

```python
dict_target={0:"setosa",1:"versicolor",2:"virginica"}
```

```python
inputarr=np.array([7.6,3.8,2.9,2.6]).reshape(1,-1)
```

```python
pred_output=df.predict(inputarr)[0]
```

```python
print(f'''The given input is:{inputarr[0]}''')
print(f'''The predicted output is : {dict_target[pred_output]}''')
```
